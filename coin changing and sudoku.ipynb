```python
''' write a python program in dynamic programmin which is the coin changing problem
example like: i gave input 3 as coins are 1,3,4 in a list and 6 as the amount to be changed
then the output should be 2 as the coins are 1,3 and the amount is 6
i need to give as user input'''

def coin_changing(coins, amount):
    dp = [0] * (amount + 1)
    dp[0] = 0
    for i in range(1, amount + 1):
        min_coins = float('inf')
        for coin in coins:
            if i - coin >= 0:
                min_coins = min(min_coins, dp[i - coin] + 1)
        dp[i] = min_coins
    return dp[amount]

coins = int(input("Enter the number of coins: "))
coin_list = []
for i in range(coins):
    coin = int(input("Enter the coin: "))
    coin_list.append(coin)
amount = int(input("Enter the amount to be changed: "))
print(coin_changing(coin_list, amount))
print(coin_changing(coin_list, amount))
print(f"The minimum number of coins to change the amount {amount} is {coin_changing(coin_list, amount)}")




print("\nOriginal Sudoku Board:")
print_board(board)
print("\nSolving...\n")

if solve_sudoku(board):
    print("Solved Sudoku Board:")
    print_board(board)
else:
    print("No solution exists for this sudoku board.")
```

```python
''' write a python program in dynamic programming of back tracking sudoku solver
but the user should give the input and he can make n*n matrix as input
and the program should be able to solve the sudoku board using back tracking algorithm'''

def solve_sudoku(board):
    empty = find_empty(board)
    if empty is None:
        return True
    row, col = empty
    n = len(board)
    for num in range(1, n + 1):
        if is_valid(board, num, row, col):
            board[row][col] = num
            if solve_sudoku(board):
                return True
            board[row][col] = 0
    return False

def find_empty(board):
    for i in range(len(board)):
        for j in range(len(board[0])):
            if board[i][j] == 0:
                return (i, j)
    return None

def is_valid(board, num, row, col):
    n = len(board)
    # Check row
    for i in range(n):
        if board[row][i] == num:
            return False
    # Check column
    for i in range(n):
        if board[i][col] == num:
            return False
    # Check box (for standard sudoku, box size is sqrt(n))
    # Only check box if n is a perfect square (like 9, 16, 25, etc.)
    box_size = int(n ** 0.5)
    if box_size * box_size == n:  # Check if n is a perfect square
        box_row = (row // box_size) * box_size
        box_col = (col // box_size) * box_size
        for i in range(box_row, box_row + box_size):
            for j in range(box_col, box_col + box_size):
                if board[i][j] == num:
                    return False
    return True

def print_board(board):
    for i in range(len(board)):
        for j in range(len(board[0])):
            print(board[i][j], end=" ")
        print()

# Get user input for board size
n = int(input("Enter the size of the sudoku board (e.g., 9 for 9x9): "))
print(f"\nEnter the {n}x{n} sudoku board (use 0 for empty cells):")
print("Enter each row separated by spaces (e.g., '5 3 0 0 7 0 0 0 0'):")

board = []
i = 0
while i < n:
    row_input = input(f"Row {i+1}: ").strip().split()
    try:
        row = [int(x) for x in row_input]
        if len(row) != n:
            print(f"Error: Row {i+1} should have {n} numbers. Please re-enter.")
            continue
        board.append(row)
        i += 1
    except ValueError:
        print(f"Error: Please enter only numbers. Please re-enter Row {i+1}.")
        continue
```
