```python
def matrix_chain_multiplication(arr):
    n = len(arr)

    # DP table
    dp = [[0 for _ in range(n)] for _ in range(n)]

    # l is chain length
    for l in range(2, n):
        for i in range(1, n - l + 1):
            j = i + l - 1
            dp[i][j] = float('inf')

            for k in range(i, j):
                cost = dp[i][k] + dp[k + 1][j] + arr[i - 1] * arr[k] * arr[j]
                dp[i][j] = min(dp[i][j], cost)

    return dp[1][n - 1]


# Driver code
arr = list(map(int, input().split(',')))
print("Minimum number of multiplications:", matrix_chain_multiplication(arr))
```
